```python
import numpy as np
class Particle(object):
  
    def __init__ (self, mass, x,y,z, vx,vy,vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
    def kinetic_e (self):
        velocity_mag= np.sqrt((self.vx**2) + (self.vy**2) + (self.vz**2))
        KE = .5*self.mass*(velocity_mag**2)
        return (KE)
    def distance_from (self, other):
        distance = np.sqrt(((other.x-self.x)**2)+((other.y-self.y)**2)+((other.z-self.z)**2))
        return (distance)
    def new_position_at(self,time):
        self.x = self.x + (self.vx*time)
        self.y = self.y + (self.vy*time)
        self.z = self.z + (self.vz*time)

    
p1 = Particle(5,6,6,6,-1,-1,-1)
p2 = Particle(2,2,4,4,-5,-8,2)
print (p1.kinetic_e())

print(p1.x,p1.y,p1.z)
p1.new_position_at(2)
print(p1.x,p1.y,p1.z)



from scipy.constants import epsilon_0
import numpy as np

K = 1 / (4 * np.pi * epsilon_0)

class ChargedParticle(Particle):
    
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
    
        super().__init__(mass, x, y, z, vx, vy, vz)
        # Add a new attribute for electric charge
        self.charge = charge
  
                
```
